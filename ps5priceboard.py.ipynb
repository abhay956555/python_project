```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = { "Year" : [2020 , 2021, 2022, 2023, 2024],
         "Price_inr": [43999, 44999, 45999, 41999, 42351,],


}

df = pd.DataFrame(data)

print(df)

sns.set_style("whitegrid")

plt.figure(figsize=(8, 9))
sns.countplot(x="Year", data=df)
plt.title("PS5 Count Plot(last 5 year)")
plt.show()

plt.figure(figsize=(8, 9))
sns.barplot(x="Year", y="Price_inr",data=df)
plt.title("PS5 price Bar Plot")
plt.ylabel("Price in inr")
plt.show()


plt.figure(figsize=(8, 9))
plt.hist(df["Price_inr"], bins=5)
plt.title("PS5 Histogram")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```
